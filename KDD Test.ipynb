## Transformações

```python
# Juntar idades em intervalos
lista = [7, 18, 99]
lista[0]
lista[2]
lista[1]

df['Pop_Edu_Ceased_Aged_15_To_17_2011'] = (df['Pop_Aged_15_And_Over_By_Age_Of_Edu_Ceased_Aged_15_2011']
                                          +df['Pop_Aged_15_And_Over_By_Age_Of_Edu_Ceased_Aged_16_2011']
                                          +df['Pop_Aged_15_And_Over_By_Age_Of_Edu_Ceased_Aged_17_2011'])
df.head()
```

```python
# Pega o nome de todas as colunas que representam uma idade específica de parada de estudos
columns_names = list(df)
aged_names = [name for name in columns_names if 'Ceased_Aged_' in name and 'Ceased_Aged_21_And_over' not in name]
aged_names
```

```python
# Exclui a coluna que criamos anteriormente
aged_names.pop()
aged_names
```

```python
# Cria novas colunas no padrão da que criamos e exclui as colunas originais
pop_aged_name = 'Pop_Aged_15_And_Over_By_Age_Of_Edu_Ceased_Aged_{}_2011'
perc_aged_name = 'Perc_Pop_Aged_15_And_Over_By_Age_Of_Edu_Ceased_Aged_{}_2011'

pop_15_to_17_2011 = 'Pop_Edu_Ceased_Aged_15_To_17_2011'
pop_18_to_20_2011 = 'Pop_Edu_Ceased_Aged_18_To_20_2011'
perc_15_to_17_2011 = 'Perc_Edu_Ceased_Aged_15_To_17'
perc_18_to_20_2011 = 'Perc_Edu_Ceased_Aged_18_To_20'

to_be_removed = []
number_of_rows = len(df)

df[pop_15_to_17_2011] = pd.Series(np.array([0]*number_of_rows))
df[perc_15_to_17_2011] = pd.Series(np.array([0]*number_of_rows))
for n in range(15, 18):
    df[pop_15_to_17_2011] = df[pop_15_to_17_2011] + df[pop_aged_name.format(n)]
    to_be_removed.append(pop_aged_name.format(n))
    df[perc_15_to_17_2011] = df[perc_15_to_17_2011] + df[perc_aged_name.format(n)]
    to_be_removed.append(perc_aged_name.format(n))
    
df[pop_18_to_20_2011] = pd.Series(np.array([0]*number_of_rows))
df[perc_18_to_20_2011] = pd.Series(np.array([0]*number_of_rows))
for n in range(18, 21):
    df[pop_18_to_20_2011] = df[pop_18_to_20_2011] + df[pop_aged_name.format(n)]
    to_be_removed.append(pop_aged_name.format(n))
    df[perc_18_to_20_2011] = df[perc_18_to_20_2011] + df[perc_aged_name.format(n)]
    to_be_removed.append(perc_aged_name.format(n))
    
print("Columns to be removed: ", to_be_removed)
```

```python
df = df.drop(to_be_removed, axis=1)
df = df.drop(['CODE'], axis=1)
df.head()
```
